```python
import numpy as np
from numpy import random
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
N = 1000
incremento = 1
X = np.zeros(N)
Y = np.zeros(N)
```

```python
for j in xrange(1, N):
    if random.uniform() > 0.5:
        X[j] = X[j-1] + incremento*2*(random.uniform() < 0.5)-1
        Y[j] = Y[j-1]
    else:
        X[j] = X[j-1]
        Y[j] = Y[j-1] + incremento*2*(random.uniform() < 0.5)-1
    
```

```python
plt.plot(X,Y)
plt.plot(X[0], Y[0], 'go', X[-1], Y[-1], 'ro')
```

Movimiento Browniano.

```python
N = 100000
T = 1.
delta_t = T/N

W_X = np.zeros(N)
W_Y = np.zeros(N)
```

```python
for j in xrange(1, N):
    if random.uniform() > 0.5:
        W_X[j] = W_X[j-1] + np.sqrt(delta_t)*(2*(random.uniform() < 0.5)-1)
        W_Y[j] = W_Y[j-1]
    else:
        W_X[j] = W_X[j-1]
        W_Y[j] = W_Y[j-1] + np.sqrt(delta_t)*(2*(random.uniform() < 0.5)-1)
```

```python
plt.plot(W_X,W_Y)
plt.plot(W_X[0], W_Y[0], 'go', W_X[-1], W_Y[-1], 'ro')
```
